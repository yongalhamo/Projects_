Net cash flow = Net operations income - monthly mortgage


net operations income = rent - expenses


expenses = vacancy rate,taxes,property managament fee

```python
# #determine Expense
# term = 30
# interest_rate=0.04
# mortgage = 100000
# down_payment = 0.20
closing_cost = 2500
# vacancy = 0.10
# maintenance = 0.05 #of rent
# management = 0.10
# annual_taxes = 1200
# rent =1000
```

```python
#This function takes listing price , mortgage term and interest rate as input and calculates the monthly motgage due

def monthly_mortgage():
    
    payment =  (interest_rate/12) * (1/(1-(1+interest_rate/12)**(-(term*12)))*mortgage)
    payment = round(payment,2)
    
    return(payment)
    
```

```python
monthly_mortgage()
```

```python
def net_cash_flow():
    
    net_operating_income = rent - ( (rent*vacancy)+ maintenance + (annual_taxes/12) + (management *rent ))
    mortgage_payment = monthly_mortgage()
    ncf = net_operating_income - mortgage_payment
    
    return (round(ncf,2))
```

```python
def cashoncash():
    annual_cashflow = net_cash_flow() * 12
    actual_cash_invested = (0.20 * mortgage) + closing_cost
    coc =  round((annual_cashflow/actual_cash_invested) * 100,1)
    return coc
```

```python
cashoncash()
```

```python
def cap_rate():
    rate = (12* (net_cash_flow() + monthly_mortgage())/mortgage) * 100 
    return rate
    
```
